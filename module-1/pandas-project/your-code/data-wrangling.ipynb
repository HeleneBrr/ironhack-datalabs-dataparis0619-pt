```python
shark = shark.rename(columns={'Case Number':"Casenumber"})
```

```python
cols = ['Casenumber', 'Case Number.1']
# applymap car on l'applique sur deux colonnes, sinon map
shark[cols] = shark[cols].applymap(lambda s: s.replace('..', '.'))
shark[cols] = shark[cols].applymap(lambda s: s.replace(',', '.'))
shark[cols] = shark[cols].applymap(lambda s: s.replace('-', '.'))
#shark[cols] = shark[cols].applymap(lambda s: s.replace('.R', ''))
# si on a R dans Casenumber alors on a Reported dans Date


# Split after replace => 2 - 3 or 4 groups
shark["splitted"] = shark["Casenumber"].map(lambda x :x.split('.'))

## Vérifier que si ND est dans l alors len(l)=2

#### aller vérifier que tous les lists sont parmi : 
# "dddd", "dd", "dd", "w"
# "dddd", "dd", "dd"
# "ND", "dddd" = "ND" et le d

def get_date_from_splitted(l):
    if (len(l)==2) & ("ND" in l) :
        return ""
    elif len(l)==3:
        return "-".join(l)
    elif len(l)==4:
        return "-".join(l[:3])
    
def get_letter_from_splitted(l):
    if len(l)==4:
        return l[3]
    else:
        return ""
 # dans map toujours le nom d'une fonction qui prend ce que tu as dans la colonne    
shark["date"] = shark["splitted"].map(get_date_from_splitted)
shark["letter"] = shark["splitted"].map(get_letter_from_splitted)
```
